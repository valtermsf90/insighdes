```python

# PASSO 1: ABIRIR BASE DE DADOS+  do kaggle
#pip install pandas openpyxl nbformat ipykernel plotly
import pandas as pd 
tabela = pd.read_csv('cancelamentos.csv')
display(tabela)


```

```python
# PASSO 2: VISUALIZAR BASE DE DADOS
      # ENTENDER AS INFORMAÇÕES DISPONÍVEIS
      # POSSIVEIS PROBLEMAS/ERROS NA BASE DE DADOS
      # inforamçoes que nao ajudam, atralham a analise

tabela = tabela.drop(columns='CustomerID')
display(tabela)

```

```python
# PASSO 3: CORRRIGIR BASE DE DADOS
display(tabela.info())
# CORRIGIR PROBLEMAS/ERROS NA BASE DE DADOS

# TRATAR VALORES FALTANTES

# TRATAR VALORES DUPLICADOS

# TRATAR OUTLIERS 

# informacoe vazias
tabela =  tabela.dropna() # eliminar linhas com valores vazios

# informacoes dados errados

display(tabela.info())


```

```python
# PASSO 4: ANALISAR BASE DE DADOS
# quantidade de cancelamentos '0' e '1'
display(tabela["cancelou"].value_counts())

# porcentagem de cancelamentos
display((tabela["cancelou"].value_counts(normalize=True)*100).map("{:.1f}%".format))


# tipo de contrato
display(tabela["duracao_contrato"].value_counts())

# porcentagem de cancelamentos por contrato 
display((tabela.groupby("duracao_contrato")["cancelou"].value_counts(normalize=True)*100).map("{:.1f}%".format))
```

```python
# passo 5 - criar gráficos para facilitar a visualização dos dados

import plotly.express as px

# criar grafico
grafico = px.histogram(tabela, x="duracao_contrato", color="cancelou")

# exibir grafico
grafico.show()
```
